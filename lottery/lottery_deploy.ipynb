```python
from web3 import Web3

w3 = Web3(Web3.HTTPProvider("http://127.0.0.1:8545"))

assert w3.is_connected(), "❌ 無法連線到 geth RPC"

print("✅ Connected to chain")
print("Chain ID:", w3.eth.chain_id)
print("Accounts:", w3.eth.accounts)
```

```python
import json

with open("build/Lottery.abi") as f:
    abi = json.load(f)

with open("build/Lottery.bin") as f:
    bytecode = f.read().strip()
```

```python
lottery = w3.eth.contract(abi=abi, bytecode=bytecode)

account = w3.eth.accounts[0] # passwd: nycu
# make sure to unlock the account in geth
# personal.unlockAccount(eth.accounts[0], "nycu", 0), 
# eth.accounts[0] will be the owner of the contract
```

```python

tx_hash = lottery.constructor().transact({
    "from": account,
    "gas": 5000000,
    "gasPrice": w3.to_wei('50', 'gwei')
})

print("Deploy tx:", tx_hash.hex())

tx_receipt = w3.eth.wait_for_transaction_receipt(tx_hash)

print("status:", tx_receipt.status)
print("Make sure to save the contract address and paste it into the lottery_frontend.ipynb!")
print("contractAddress:", tx_receipt.contractAddress)
```
