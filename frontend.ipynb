```python
from web3 import Web3
import json

w3 = Web3(Web3.HTTPProvider("http://127.0.0.1:8545"))
assert w3.is_connected()

with open("build/SimpleStorage.abi") as f:
    abi = json.load(f)

contract_address = Web3.to_checksum_address("0x68443Ae855bE6f3468093237510AEBA145b07bE2")

storage = w3.eth.contract(
    address=contract_address,
    abi=abi
)
```

```python
code = w3.eth.get_code(contract_address)
print(code.hex())
```

```python
account = w3.eth.accounts[0]

tx_hash = storage.functions.set(100).transact({
    "from": account,
    "gas": 100_000
})

receipt = w3.eth.wait_for_transaction_receipt(tx_hash)
print("set() tx mined in block:", receipt.blockNumber)
```

```python
value = storage.functions.get().call()
print("storedData =", value)
```
